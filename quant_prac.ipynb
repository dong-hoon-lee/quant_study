```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = './data/'

df = pd.read_csv(data_path+'samsung.csv')
df
```

```python
df.info()
```

```python
def date_transform(df):
    df['Date'] = pd.to_datetime(df['Date'])
    return df
```

```python
date_transform(df)
```

```python
df.set_index('Date', inplace=True)
df
```

```python
df['2021-06']
```

```python
# shift 함수
# 데이터에 시차를 줌
# default period=1

df['Close_lag_1'] = df['Close'].shift()
df
```

```python
# pct_change 함수
# = percentage change
# 말 그대로 how much the value has changed in percentage

df['pct_change'] = df['Close'].pct_change()
df
```

```python
# diff 함수
# diff_1 = value_2 - value_1
# 즉, 현재값 - 이전값

df['Close_diff'] = df['Close'].diff()
df
```

```python
# rolling 함수
# 이동평균 구해줌 (mean of N days)

df['MA'] = df['Close'].rolling(window=5).mean()    # 5-days mean
df
```

```python
print(type(df['MA']))
```

```python
# resample
# 금융 데이터는 upsampling 보다 downsampling을 더 많이 사용함

# resample 찾아보기!!!

idx = pd.date_range(start='2021-05', end='2021-10')
series = pd.Series(range(len(idx)), index=idx)
series
```

```python
series.resample(rule='M').sum()
```

```python
series.resample(rule='M').last()
```

```python
# daily return(일별 수익률) 계산

df['daily_rt'] = df['Adj Close'].pct_change()
df
```
